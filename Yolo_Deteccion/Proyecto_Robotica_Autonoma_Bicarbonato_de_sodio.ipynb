```python
import zipfile
import os
# Ruta al archivo ZIP en tu Google Drive
zip_path = '/content/drive/MyDrive/Proyecto_Robotica_Autonoma/BICAR.v1i.yolov8.zip'

# Ruta donde quieres extraer los contenidos del ZIP
extract_path = '/content/data_ra'
```

```python
# Crear el directorio si no existe
if not os.path.exists(extract_path):
    os.makedirs(extract_path)
# Extraer el contenido del archivo ZIP
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
modelo = YOLO("yolov8x.pt")
```

```python
modelo.train(data="/content/data_ra/data.yaml", epochs=30)
```

```python
modelo2 = YOLO("runs/detect/train/weights/best.pt")
```

```python
resultados2 = modelo2.predict("/content/data_ra/valid/images/IMG_20240719_093651_jpg.rf.59ec20b2d5ca1d80f57e75c315dfb0aa.jpg", conf=0.7)
```

```python
# Imagen resultante (con recuadros delimitadores)
I = resultados2[0].plot()

# Mostrar la figura
plt.imshow(I[:,:,::-1])
plt.axis('off');
```
